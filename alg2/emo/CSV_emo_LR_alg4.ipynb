```python
import numpy as np
import warnings
warnings.filterwarnings("ignore")

X_text_train = np.load("/Users/eciel/Desktop/y3s2_sum/orf/summer/privacy_methods/LR/IF/Logistic_emotion/X_train_text.npy")
X_text_dev = np.load("/Users/eciel/Desktop/y3s2_sum/orf/summer/privacy_methods/LR/IF/Logistic_emotion/X_test_text.npy")
```

```python
y={}
y["dev"] = np.load("/Users/eciel/Desktop/y3s2_sum/orf/summer/privacy_methods/LR/IF/Logistic_emotion/y_dev.npy")
y["train"] = np.load("/Users/eciel/Desktop/y3s2_sum/orf/summer/privacy_methods/LR/IF/Logistic_emotion/y_train.npy")
y["dev"].shape
```

```python
app_k = np.array(np.load("NT_app_k_emo_LR_I100_D1.npy", allow_pickle=True).squeeze())
pred = np.load("/Users/eciel/Desktop/y3s2_sum/orf/summer/privacy_methods/LR/IF/Logistic_emotion/old_predictions.npy", allow_pickle=True)
app_k_nonon=[]
for i in range(X_text_dev.shape[0]):
    print(app_k[i])
    if app_k[i] == None:
        app_k_nonon.append(X_text_train.shape[0])
    else:
        app_k_nonon.append(app_k[i])
```

```python
sort_dev_k = np.array(app_k_nonon, dtype=int).argsort()
sort_dev_k
```

```python
new_predictions = np.load("new_predictions_emo_LR_I100_D1.npy", allow_pickle=True)
```

```python
data = []
for test_movie_id in sort_dev_k:
    
    test_point = []
    # id
    test_point.append(test_movie_id)
    # test_movie_review_text
    test_point.append("label: " + str(y["dev"][test_movie_id]) + "  doc: " + str(X_text_dev[test_movie_id]))
    test_point.append(pred[test_movie_id].item())
    test_point.append(np.round(pred[test_movie_id].item()) == y["dev"][test_movie_id].item())
    # k
    k = app_k[test_movie_id]
    test_point.append(str(k))
    print("k", k)
    if k == None:
        test_point.append(None)
        data.append(test_point)
        continue
    k = int(k)
    test_point.append(np.round(new_predictions[test_movie_id]) != np.round(pred[test_movie_id].item()))
    top_k_index = order_lis[test_movie_id]
    print("k", k)
    
    # training examples
    for idx in top_k_index:
        test_point.append("id: " + str(idx) + "  label: " + str(y["train"][idx]) + "  doc: " + str(X_text_train[idx]))
    data.append(test_point)
```

```python
import csv

with open('CSV_emo_LR_alg4.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(header)

    # write multiple rows
    writer.writerows(data)
```
